**Question 2 – European Used Cars Body Type Classification**

In this task, I analyze a cleaned Dutch used-cars dataset containing 104,426 listings
and 32 structured features. The objective is to classify the vehicle BodyType
(Hatchback, SUV, Sedan, MPV, Coupe, etc.) using technical specifications,
usage information, and TF-IDF equipment embeddings (A1–A4).

The workflow includes data understanding, exploratory analysis, preprocessing,
feature engineering, machine-learning modeling, and evaluation using
classification metrics.


# Step 1: Import Libraries
I used these Python libraries:
- `pandas` and `numpy` for data
- `matplotlib` and `seaborn` for plots
- `sklearn` for machine learning and metrics


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score,f1_score ,classification_report,confusion_matrix

import joblib
import warnings
warnings.filterwarnings("ignore")
```

# Step 2: Load Data
I read my CSV file `used_cars.csv` and looked at it.
- `data.info()` to see columns and missing values
- `data.describe()` to see numbers
- `data.head()` to see first rows
- I checked `BodyType` counts to know car types


```python
data=pd.read_csv("used_cars.csv",low_memory=False)
print(data.info())
print(data.describe())
print(data['BodyType'].value_counts())
print(data.head())
print(data.shape)
```

# Step 3: Plot BodyType
I made a bar chart to see which car BodyType is more.


```python
data['BodyType'].value_counts().plot(
    kind='bar',
    figsize=(6,4),
    color='skyblue'
)
plt.title("BodyType Distribution")
plt.xlabel("BodyType")
plt.ylabel("Count")
plt.show()
```

# Features & Missing Values
- Fill missing numbers with median.  
- Fill missing categories with "Unknown".  
- New: VehicleAge, Power_Weight_Ratio, Log_Km.


```python
cat_cols = ['Brand','Model','FuelType','Transmission','Drive']
num_cols = ['Power_kW_HP','EngineCapacity','EmptyWeight','CO2Emission','Gears','Seats','Km']

for col in num_cols:
    data[col] = data.groupby(['Brand','Model','YearBuilt'])[col] \
                    .transform(lambda x: x.fillna(x.median()))

for col in cat_cols:
    data[col] = data[col].fillna("Unknown")

data['VehicleAge'] = 2026 - data['YearBuilt'] 
data['Power_Weight_Ratio'] = data['Power_kW_HP'] / data['EmptyWeight']
data['Log_Km'] = np.log1p(data['Km']) 

num_cols += ['VehicleAge','Power_Weight_Ratio','Log_Km']
```

# Graphs
I made 6 graphs to see the data:  
- Bar chart for BodyType counts

```python
plt.figure(figsize=(6,4))
data['BodyType'].value_counts().plot(kind='bar', color='skyblue')
plt.title("BodyType Distribution")
plt.show()

plt.figure(figsize=(8,4))
sns.histplot(data['Price'], bins=30, kde=True, color='green')
plt.title("Price Distribution")
plt.show()

plt.figure(figsize=(8,4))
sns.scatterplot(x='YearBuilt', y='Price', data=data, hue='BodyType', alpha=0.6)
plt.title("YearBuilt vs Price")
plt.show()

plt.figure(figsize=(8,4))
sns.scatterplot(x='Km', y='Price', data=data, hue='BodyType', alpha=0.6)
plt.title("Km vs Price")
plt.show()

plt.figure(figsize=(8,4))
sns.scatterplot(x='Power_Weight_Ratio', y='Price', data=data, hue='BodyType', alpha=0.6)
plt.title("Power to Weight Ratio vs Price")
plt.show()

plt.figure(figsize=(8,4))
sns.scatterplot(x='VehicleAge', y='Price', data=data, hue='BodyType', alpha=0.6)
plt.title("Vehicle Age vs Price")
plt.show()
```

# Fill Missing Values
- For number columns, fill missing values with median per Brand+Model+YearBuilt.
- For category columns, fill missing values with "Unknown".


```python
num_cols = ['Power_kW_HP','EngineCapacity','EmptyWeight','CO2Emission','Gears','Seats','Km']
for col in num_cols:
    data[col] = data.groupby(['Brand','Model','YearBuilt'])[col]\
                   .transform(lambda x: x.fillna(x.median()))

cat_cols = ['FuelType','Transmission','Drive']
for col in cat_cols:
    data[col] = data[col].fillna("Unknown")
```

# New Features
- VehicleAge = 2026 - YearBuilt  
- Power_Weight_Ratio = Power / Weight  
- Log_Km = log(Km + 1)  
- Check first 5 rows of new features.


```python
data['VehicleAge'] = 2026 - data['YearBuilt']
data['Power_Weight_Ratio'] = data['Power_kW_HP'] / data['EmptyWeight']
data['Log_Km'] = np.log1p(data['Km'])

print(data[['VehicleAge','Power_Weight_Ratio','Log_Km']].head())
```

```python
cat_features = ['Brand','Model','FuelType','Transmission','Drive']
num_features = ['YearBuilt','Power_kW_HP','EngineCapacity','EmptyWeight',
                'CO2Emission','Gears','Seats','VehicleAge',
                'Power_Weight_Ratio','Log_Km','A1','A2','A3','A4']

X = data[cat_features + num_features]
y = data['BodyType']

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
preprocessor = ColumnTransformer([
    ('num',StandardScaler(),num_features),
    ('cat',OneHotEncoder(handle_unknown='ignore'),cat_features)
])
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X,y_encoded,test_size=0.2, stratify=y_encoded,random_state=42
)
```

```python
models={
    "LogisticRegression":LogisticRegression(max_iter=1000,class_weight='balanced'),
   "Random Forest": RandomForestClassifier(n_estimators=200, max_depth=15,
                                         random_state=42, class_weight='balanced'),
"SVM":SVC(kernel='rbf',class_weight='balanced')
}
```

# Train & Check Models
- Use pipeline: preprocess + model  
- Fit on train data, predict test data  
- Show Accuracy, F1 scores, and classification report  
- Draw confusion matrix for each model


```python
for name, model in models.items():
    pipe = Pipeline([
        ('prep', preprocessor),
        ('model', model)
    ])
    
    pipe.fit(X_train, y_train)
    preds = pipe.predict(X_test)
    
    print("Model:", name)
    
    print("Accuracy:", round(accuracy_score(y_test, preds), 4))
    print("Macro F1 Score:", round(f1_score(y_test, preds, average='macro'), 4))
    print("Weighted F1 Score:", round(f1_score(y_test, preds, average='weighted'), 4))
    
    print("\nClassification Report:")
    print(classification_report(y_test, preds, target_names=le.classes_))
    
    cm = confusion_matrix(y_test, preds)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```
