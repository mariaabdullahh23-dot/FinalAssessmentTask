## Stock Price Prediction using LSTM
Ye code Microsoft (MSFT) ka stock data read karta hai, clean karta hai, aur LSTM model use karke Close price predict karta hai. Graphs aur metrics (RMSE, MAE) bhi calculate kiye gaye hain.


#  Import Libraries


#  Load Dataset


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

#  Load Dataset & Fix Columns


```python
data = pd.read_csv("MSFT.csv")

data.columns = data.columns.str.strip()

for col in data.columns:
    if 'date' in col.lower():
        date_column = col
        break
data[date_column] = pd.to_datetime(data[date_column])
data = data.sort_values(date_column)
data.set_index(date_column, inplace=True)

print("Date column detected and set as index:", date_column)
```

# Data Info & Cleaning


```python
print(data.info())
print("Shape of data:", data.shape)
print("Null values:\n", data.isnull().sum())

data = data.drop_duplicates()
```

#  Plot Close Price & Volume


```python
plt.figure(figsize=(10,5))
plt.plot(data.index, data['close'])   
plt.title("MSFT Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()

plt.figure(figsize=(10,5))
plt.plot(data.index, data['volume']) 
plt.title("MSFT Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

# Select Features


```python
features=['open','high','low','close','volume']
dataset=data[features]
```

#  Normalize Data


```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(dataset)
```

#  Create Sequences


```python
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i])
        y.append(data[i, 3]) 
    return np.array(X), np.array(y)

WINDOW_SIZE = 30
X, y = create_sequences(scaled_data, WINDOW_SIZE)
```

#  Train / Validation / Test Split


```python
train_size = int(0.7 * len(X))
val_size = int(0.85 * len(X))

X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:val_size], y[train_size:val_size]
X_test, y_test = X[val_size:], y[val_size:]
```

#  Early Stopping


```python
early_stop =EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)
    
```

#  Define RNN, LSTM, GRU Models


```python
def build_rnn():
    model = Sequential()
    model.add(SimpleRNN(64, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

def build_lstm():
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(0.2))
    model.add(LSTM(64))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

def build_gru():
    model = Sequential()
    model.add(GRU(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(0.2))
    model.add(GRU(64))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model
```

#  Train LSTM Model


```python
model = build_lstm()

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stop],
    verbose=1
)
```

#  Plot Training vs Validation Loss


```python
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```

#  Predictions & Inverse Scaling


```python
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

close_index = 3

train_pred_full = np.zeros((len(train_pred), len(features)))
train_pred_full[:,close_index] = train_pred.flatten()
train_pred_actual = scaler.inverse_transform(train_pred_full)[:,close_index]

test_pred_full = np.zeros((len(test_pred), len(features)))
test_pred_full[:,close_index] = test_pred.flatten()
test_pred_actual = scaler.inverse_transform(test_pred_full)[:,close_index]

y_train_full = np.zeros((len(y_train), len(features)))
y_train_full[:,close_index] = y_train
y_train_actual = scaler.inverse_transform(y_train_full)[:,close_index]

y_test_full = np.zeros((len(y_test), len(features)))
y_test_full[:,close_index] = y_test
y_test_actual = scaler.inverse_transform(y_test_full)[:,close_index]
```

#  Plot Actual vs Predicted


```python
plt.figure(figsize=(10,5))
plt.plot(y_train_actual, label='Actual Train')
plt.plot(train_pred_actual, label='Predicted Train')
plt.title("Train: Actual vs Predicted")
plt.legend()
plt.show()

plt.figure(figsize=(10,5))
plt.plot(y_test_actual, label='Actual Test')
plt.plot(test_pred_actual, label='Predicted Test')
plt.title("Test: Actual vs Predicted")
plt.legend()
plt.show()
```

# Recent Period Prediction


```python
plt.figure(figsize=(10,5))
plt.plot(y_test_actual[-200:], label='Actual')
plt.plot(test_pred_actual[-200:], label='Predicted')
plt.title("Recent Period Prediction (Zoomed)")
plt.legend()
plt.show()
```

#  Calculate Metrics


```python
rmse = np.sqrt(mean_squared_error(y_test_actual, test_pred_actual))
mae = mean_absolute_error(y_test_actual, test_pred_actual)

print("RMSE:", rmse)
print("MAE :", mae)
```
